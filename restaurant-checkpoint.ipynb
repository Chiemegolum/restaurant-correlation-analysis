### Top 5 restaurants by Revenue

```python
# Top 5 restaurants by Revenue
top_5_restaurants= rest.groupby('Name')['Revenue'].mean().reset_index().sort_values(by= 'Revenue', ascending= False)
top_5_restaurants.head(5)
```

```python
# Plotting the top 5 restaurants by revenue
plt.bar(top_5_restaurants.head(5)['Name'], top_5_restaurants.head(5)['Revenue'], color= 'red')                               
plt.title('Top 5 restaurants by Revenue', fontsize= 18, fontweight= 'bold')
plt.xlabel('Name')
plt.ylabel('Revenue')
plt.xticks(rotation= 45)
plt.ticklabel_format(style= 'plain', axis= 'y')            # Prevents scientific notation on the y-axis
plt.show()
```

### Insight: 
#### The chart displays the 5 highest-earning restaurnts in the dataset. They showed high stability (ranging around 1.5 million) and minimal variance amongst them. 
#### This suggests a tight competition amongst the restaurants, indicating that minor operational improvements or targeted marketing campaigns by any of these competitors could easily shift the market leadership. Further analysis into customer retention rates or transaction volumes is required to determine what gives Restarant 4324 its slight edge.

### 5 most Rated restaurants

```python
# Checking how many restaurants were rated highest (5)
five_star= rest[rest['Rating']== 5]
five_star
```

### Since there are lots of restaurants rated 5, I will use the 'Number of Reviews' to show the most popular amongst the top-tier restaurants.

```python
five_most_popular= five_star.nlargest(5, 'Number of Reviews')
five_most_popular
```

```python
# Plotting the 5 most popular restaurants

plt.barh(five_most_popular['Name'], five_most_popular['Number of Reviews'].sort_values(), color= 'red')                               
plt.title('Top 5 Most Popular 5-Star Restaurants', fontsize= 18, fontweight= 'bold')
plt.xlabel('Number of Reviews')
plt.ylabel('Name', rotation= 0, labelpad= 30)

# Zoom in on x-axis to make difference visible
plt.xlim(950, 990)

plt.show()
```

### Insight:
#### The graph clearly shows that restaurant 17 was the most popular, followed by the restaurant 3418.

### Carry out EDA to find the relation between revenue and features of the dataset.

```python
# Finding any duplicates
duplicate_rows_rest= rest[rest.duplicated()]
print('number of duplicate rows:', duplicate_rows_rest.shape)
```

```python
# Plotting different features against one another

plt.figure(figsize= (15,8))

c= rest.select_dtypes(include= 'number').corr()

sns.heatmap(c, cmap= 'coolwarm', annot= True)
plt.title('Feature Correlation Matrix(Focus on Revenue)', fontsize= 18, fontweight= 'bold')
plt.xticks(rotation= 45)
plt.tight_layout()
```

### Insight:
#### The 2 clear standouts: average meal price(0.69), and seating capacity(0.68) seem to be the primary drivers of revenue with their strong positive correlation. They indicate that increase in average meal price and the ability to seat more customers is heavily tied to higher overall earnings.

#### The marketing budget(0.37), and social media followers(0.35) seem to be the secondary revenue drivers as they show moderate positive correlation. Investing in marketing and growing a digital audience improves revenue, tough less drastically than pricing and capacity.

#### Next to these are the weekend reservations(0.29), and weekday reservations(0.27) with very weak positive link to revenue.

#### Showing little or no correlation at all to the revenue are the: ratings, chef experience years, ambience score, service quality score, number of reviews, and average review length. These explain that a restaurant with high ratings or perfect quality service scores do not imply that it generates huge income. Thus, they do not have any relationship with revenue coming in.
